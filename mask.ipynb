```python
import nibabel as nib
import numpy as np
from nibabel.processing import resample_from_to

# Load images
nii_img1 = nib.load(nii_file_path1)
nii_img2 = nib.load(nii_file_path2)
nii_img4 = nib.load(nii_file_path4)
nii_img5 = nib.load(nii_file_path5)

# Print affine matrices
print("Affine matrix of nii_img1:")
print(nii_img1.affine)

print("\nAffine matrix of nii_img2:")
print(nii_img2.affine)

print("\nAffine matrix of nii_img4:")
print(nii_img4.affine)

print("\nAffine matrix of nii_img5:")
print(nii_img5.affine)

# Define reference (e.g., nii_img2)
reference = (nii_img2.shape, nii_img2.affine)

# Resample images to reference if necessary
resampled_img1 = resample_from_to(nii_img1, reference)
resampled_img4 = resample_from_to(nii_img4, reference)
resampled_img5 = resample_from_to(nii_img5, reference)

# Get data from images
img_data1 = resampled_img1.get_fdata()
img_data2 = nii_img2.get_fdata()  # Already in reference space
img_data4 = resampled_img4.get_fdata()
img_data5 = resampled_img5.get_fdata()

# Create mask
mask = np.zeros(reference[0])
mask[img_data1 > 0] = 1
mask[img_data2 > 0] = 1
mask[img_data4 > 0] = 1
mask[img_data5 > 0] = 1

# Save the mask using the affine matrix from the reference image
nib.save(nib.Nifti1Image(mask, nii_img2.affine), 'FP_Mask_Final.nii')
```
